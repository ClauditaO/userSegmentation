```python
import codecs
import os
import numpy as np
```

```python
#@inproceedings{sap2014developing,
# author={Sap, Maarten and Park, Greg and Eichstaedt, Johannes C and Kern, Margaret L and Stillwell, David J and Kosinski, Michal and Ungar, Lyle H and Schwartz, H Andrew},
# title={Developing age and gender predictive lexica over social media},
# booktitle={Proceedings of the 2014 Conference on Empirical Methods in Natural Language Processing (EMNLP)},
# year={2014},
# }
# The lexicon I used is in: http://www.wwbp.org/lexica.html#refinement

intercept = -0.06724152
lexicon = dict()
no_weights_lexicon = dict()

dirname = '/home/clau/Documents/journal_five_countries/'

for file in os.listdir(dirname+'/audience_tweets_ireland/'):
    lexicon_file = codecs.open('/home/clau/userSegmentationPaper/resources/genderLexicon.txt')
    for line in lexicon_file:
        word_weight = line.split('\t')
        lexicon[word_weight[0]] = float(word_weight[1].strip())
        no_weights_lexicon[word_weight[0]] = 0
    lexicon_file.close()
    #print(file)
    this_user_lexicon = no_weights_lexicon
    this_user_tweets = codecs.open(dirname+'/audience_tweets_ireland/'+file, 'r', encoding='utf-8')
    tweets_text = ''
    for line in this_user_tweets:
        pieces = line.split('\t')
        tweets_text = tweets_text+' '+pieces[0].replace('\n\n','\n')
    users_tokens = tweets_text.split(' ')
    total_words_used = len(users_tokens)
    #print(total_words_used)
    for token in users_tokens:
        if token in this_user_lexicon:
            this_user_lexicon[token] += 1
    lexicon_usage = 0
    for k,v in this_user_lexicon.items():
        if v > 0:
            #print(k)
            lexicon_usage = lexicon_usage + ((v/total_words_used)*lexicon[k])
    if lexicon_usage + intercept > 1:
        print(file.replace('.txt','')+'\tfemale')
    elif np.sign(lexicon_usage + intercept) < 0:
        print(file.replace('.txt','')+'\tmale')
    else:
        print(file.replace('.txt','')+'\tnone')
```

```python
from sklearn.metrics import classification_report

#best F1-score achieved when I use the genderizer, and for the cases when the genderizer returns None I use the lexicon generated gender
#furthermore, when the lexicon returns None, I use Female. F1-score: 0.94
#The true labels are labels I manually assigned to journalists

y_true = [2,2,2,3,2,2,1,4,2,3,3,2,2,2,3,2,4,1,2,3,2,1,2,2,3,3,4,2,1,1,1,2]
y_pred = [2,2,2,3,2,2,1,4,2,3,3,2,1,2,3,2,4,1,3,4,2,1,2,2,3,3,2,2,1,1,4,1]
target_names = ['class 1', 'class 2','class 3', 'class 4']
print(classification_report(y_true, y_pred, target_names=target_names))
```
