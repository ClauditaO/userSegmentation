```python
import re
import requests
from bs4 import BeautifulSoup
import codecs
```

```python
names_file = codecs.open('/home/clau/Documents/journal_five_countries/journos_names.txt')
for line in names_file:
    journo_name = ''
    name = line.strip().split(' ')
    for n in name:
        journo_name = journo_name+' '+n
        #print(journo_name.strip().replace(' ','_'))
    url = 'https://en.wikipedia.org/w/api.php?action=query&prop=revisions&rvprop=content&rvsection=0&titles='+journo_name.strip().replace(' ','_')+'&format=xml'

    res = requests.get(url)
    soup = BeautifulSoup(res.text, "xml")

    if "-1" not in str(soup.pages) and '#REDIRECT' not in soup.revisions.getText() and 'may refer to' not in soup.revisions.getText():
        birth_re = re.search(r'(Birth_date(.*?)}})', soup.revisions.getText())
        birth_data = 'none'
        if birth_re is not None:
            if "<!--" in str(birth_data):
                birth_data = 'none'
            else:
                birth_data = birth_re.group(0)
    else:
        birth_data = 'none'

    print(journo_name+'\t'+str(birth_data))
names_file.close()
```
