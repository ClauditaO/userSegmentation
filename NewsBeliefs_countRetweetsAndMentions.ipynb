```python
import codecs
import os
```

```python
path = '/home/clau/Documents/journalists2017_personas/'
journos_file = codecs.open(path+'journos_who_tweeted.txt','r')
journos_screennames = list()
for line in journos_file:
    journos_screennames.append(line.strip())
journos_file.close()
print('user_id\tretweets_percent\tmentions_percent\ttotal_interactions')
for f in os.listdir(path+'audience_tweets_ireland/'):
    print(f)
    unique_tweets = set()
    this_user_file = codecs.open(path+'audience_tweets_ireland/'+f, 'r')
    retweets_f = codecs.open(path+'audience_retweets_ireland/'+f, 'w')
    mentions_f = codecs.open(path+'audience_mentions_ireland/'+f, 'w')
    retweets = 0
    mentions = 0
    for line in this_user_file:
        for js in journos_screennames:
            if js in line and line.strip() not in unique_tweets:
                unique_tweets.add(line.strip())
                if "RT @" in line:
                    retweets += 1
                    #retweets_f.write(line.strip()+'\n')
                else:
                    mentions += 1
                    #mentions_f.write(line.strip()+'\n')
    retweets_f.close()
    mentions_f.close()
    this_user_file.close()
    if retweets+mentions > 0:
        print(f.replace(".txt", "")+'\t'+str(retweets/float(retweets+mentions))+'\t'+str(mentions/float(retweets+mentions))+'\t'+str(retweets+mentions))
    else:
        print(f.replace(".txt", "")+'\t0\t0\t0')
```
