```python
import codecs
import os
from collections import Counter
import re
```

```python
#News Motivations
#21005 interactions only with individual journalists
#37076 interactions only with corporate accounts
#printing order: lifestyle, sports, breaking_news, corporate, politics, sci&tech, business
path = '/home/clau/Documents/journalists2017_personas/'
#load journalists' categories
journos_file = codecs.open(path+'irish_journos_info.txt','r')
journos_categories = dict()
for line in journos_file:
    parts = line.split('\t')
    journos_categories[parts[3]] = parts[4].strip()
#load users' interactions' counts [retweets+mentions]
#fields: user_id,retweets_percent,mentions_percent,total_interactions
audience_interactions_file = codecs.open(path+'retweets_mentions_proportion.txt','r')
user_interactions = dict()
total_corporate = 0
for line in audience_interactions_file:
    retweets = 0
    mentions = 0
    categories_interactions = Counter()
    categories_interactions['business'] = 0
    categories_interactions['lifestyle'] = 0
    categories_interactions['breaking_news'] = 0
    categories_interactions['politics'] = 0
    categories_interactions['sci&tech'] = 0
    categories_interactions['sports'] = 0
    categories_interactions['corporate'] = 0
    user_parts = line.split('\t')
    if float(user_parts[1]) > 0:
        user_retweets_file = codecs.open(path+'/audience_retweets_ireland/'+user_parts[0]+'.txt','r')
        for retweet in user_retweets_file:
            journos_present = re.findall(r'@(\w+)', retweet)
            for journo in journos_present:
                if journo.replace(':', '') in journos_categories.keys():
                    categories_interactions[journos_categories[journo]] += 1
    if float(user_parts[2]) > 0:
        user_mentions_file = codecs.open(path+'/audience_mentions_ireland/'+user_parts[0]+'.txt','r')
        for mention in user_mentions_file:
            journos_present = re.findall(r'@(\w+)', mention)
            for journo in journos_present:
                if journo.replace(':', '') in journos_categories.keys():
                    categories_interactions[journos_categories[journo].strip()] += 1
    to_print = ''
    total = 0
    for c,v in categories_interactions.items():
        total = total+v
    if total > 0:
        for c,v in categories_interactions.items():
            to_print = to_print+str(v/float(total))+','
        print(user_parts[0]+','+to_print)
#    total_corporate = total_corporate+categories_interactions['corporate']
#print(total_corporate)
```
